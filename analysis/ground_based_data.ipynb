# First inspection of ground-based data sets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## load and inspect datasets

```python
d1 = pd.read_csv("../datasets/ch4_gc-md_monthly.csv")
d2 = pd.read_csv("../datasets/monthly_ch4_in-situ.csv")
d3 = pd.read_csv("../datasets/monthly_ch4_surface-flask.csv")
d4 = pd.read_csv("../datasets/monthly_co2_in-situ.csv")
d5 = pd.read_csv("../datasets/monthly_co2_surface-flask.csv")

# inspect data set
d4
```

```python
print(d2.columns)
print(d3.columns)
print(d4.columns)
print(d5.columns)
```

## plot data

```python
_, axs = plt.subplots(2,1, figsize=(7,6), sharex=True)
cols = sns.color_palette("hls", 3)

for i, (d,l) in enumerate(zip([d1, d2, d3], ["gc", "in-situ","surface-flask"])):
    sns.lineplot(data=d, x="year", y="value", ax=axs[0], color=cols[i],
                 label=l)
for i, (d,l) in enumerate(zip([d4, d5], ["in-situ","surface-flask"])):
    sns.lineplot(data=d, x="year", y="value", ax=axs[1], color=cols[i],
                 label=l)
[axs[i].legend(title="method", frameon=False, handlelength=0.3) for i in range(2)]
axs[0].set_xlabel("year")
axs[0].set_title(r"$CH_4$")
axs[1].set_title(r" $C0_2$")
plt.show()
```
